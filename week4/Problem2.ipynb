```python
test=pd.read_csv('sentimentanalysis_test.csv',encoding="latin-1")
train=pd.read_csv('sentimentanalysis_train.csv',encoding="latin-1")
processed_train=train.copy()
processed_test=test.copy()
processed_test['Sentence']=test['Sentence'].str.lower()
processed_train['Sentence']=train['Sentence'].str.lower()
```

```python
from nltk import word_tokenize,pos_tag

def lemmatization(text):
    result=[]
    wordnet = WordNetLemmatizer()
    for token,tag in pos_tag(text):
        pos=tag[0].lower()
        if pos not in ['a', 'r', 'n', 'v']:
            pos='n'
            
        result.append(wordnet.lemmatize(token,pos))
    
    return result
tex = ['running','ran','runs'] 

lemmatization(tex)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords_lower = [s.lower() for s in stopwords]
stopwords_lower
```

```python
def text_preprocessing(str_input): 
     #tokenization, remove punctuation, lemmatization
     t=[word for word in word_tokenize(str_input) if word.isalpha()]
     words=lemmatization(t)
     # remove symbols, websites, email addresses 
     words = [re.sub(r"[^A-Za-z@]", "", word) for word in words] 
     words = [re.sub(r"\S+com", "", word) for word in words]
     words = [re.sub(r"\S+@\S+", "", word) for word in words] 
     words = [word for word in words if word!=' ']
     words = [word for word in words if len(word)!=0] 
 
     #remove stopwords     
     words=[word.lower() for word in words if word.lower() not in stopwords_lower]
     #combine a list into one string   
     string = " ".join(words)
     return string
 
```

```python
text_preprocessing(text)
```

```python
processed_test['Sentence']=test['Sentence'].apply(text_preprocessing)
processed_train['Sentence']=train['Sentence'].apply(text_preprocessing)
```
